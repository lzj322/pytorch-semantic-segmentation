## 读图写图

```python
from PIL import Image
%matplotlib inline
import matplotlib.pyplot as plt
import torch
import numpy as np
import cv2
```

这里是palette 来给数字标号的图像一个调色板

```python
palette = [0, 0, 0, 128, 0, 0, 0, 128, 0, 128, 128, 0, 0, 0, 128, 128, 0, 128, 0, 128, 128,
           128, 128, 128, 64, 0, 0, 192, 0, 0, 64, 128, 0, 192, 128, 0, 64, 0, 128, 192, 0, 128,
           64, 128, 128, 192, 128, 128, 0, 64, 0, 128, 64, 0, 0, 192, 0, 128, 192, 0]
```

```python
def colorize_mask(mask):
    # mask: numpy array of the mask
    new_mask = Image.fromarray(mask.astype(np.uint8)).convert('P')
    new_mask.putpalette(palette)

    return new_mask
```

```python
voc_name = '2007_000033.png'
```

```python
lip_name = '328_470613.png'
```

```python
im_voc = Image.open(voc_name)
im_lip = Image.open(lip_name)
```

```python
plt.imshow(im_voc)
plt.show()
plt.imshow(im_lip)
```

将图像转成numpy 之后，就可以用

```python
im_mask = colorize_mask(np.array(im_lip))
# im_mask2 = im_lip.putpalette(palette)
```

```python
plt.imshow(im_mask)
# plt.imshow(im_mask2)
```

```python
im_mask.save('image_palette.png')
```
